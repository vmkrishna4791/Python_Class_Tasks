```python
#Down load the csv file called carbon nano tubes from UCI repository
#create table if not existing.
#Read the contents row by row, do processing and segreagate the data as instructed by UCI carbo nano tubes link.
#Insert the processed data into the Carbon tube table in local mysql database called student.

import os
import mysql.connector as connection
import csv



try:
    
    mydb = connection.connect(host="localhost",database='student',user="root", passwd="mysql",use_pure=True)

    mycursor = mydb.cursor()

    mycursor.execute("Show tables;")

    myresult = mycursor.fetchall()


#check to see if the table exists? If yes, it is dropped and created again

    for table_name in myresult:
            if (str(table_name[0])=='carbontubes'):
                mycursor.execute('drop table carbontubes')

#table creation

    query = "CREATE TABLE carbontubes (Cin varchar(10), Cim varchar(10), Iacu varchar(60), Iacv varchar(60), Iacw varchar(60), Cacu varchar(60), Cacv varchar(60), Cacw varchar(100))"

    mycursor.execute(query)
   
    
    count=0 
    
    with open('carbon_nanotubes.csv',"r") as data:
        next(data)
        data_csv=csv.reader(data,delimiter='\n')
#        print('Record:',data_csv)
 
        for i in (data_csv):
            lst=[]

# The record from csv file processed to replace the semicolon with space and also removing characters like '[',']','\' 
            val=str(i).replace(';',' ').replace('[','').replace(']','').replace('\'','')
            
# Further splitting the above record based on comma seperation into a list called Var
            var=val.split(' ')
    
#converting the entire list var into a single string and appending into a list called lst
            
                        
            new_str1='\"'+var[0]+'\"'+","+'\"'+var[1]+'\"'+","+'\"'+var[2]+'\"'+","+'\"'+var[3]+'\"'+","+'\"'+var[4]+'\"'+","+'\"'+var[5]+'\"'+","+'\"'+var[6]+'\"'+","+'\"'+var[7]+'\"'    
            
            
            mycursor.execute("insert into student.carbontubes values ({record})".format(record=(new_str1)))
            mydb.commit()
            
            count=count+1
            print("No of records inserted:",count)
                  
    mydb.close()      
                

except Exception as e:
    print(str(e))
    mydb.close()
```
